```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
china1965 = pd.read_csv('income-1965-china.csv')
usa1965 = pd.read_csv('income-1965-usa.csv')
china1965.head()
```

```python
china1965.income.plot(kind = 'box')
```

```python
pd.DataFrame({'usa':usa1965.income, 'china':china1965.income}).boxplot()
```

```python
_ = plt.hist(usa1965, bins = 10)
_ = plt.xlabel('income')
_ = plt.ylabel('Ylabel')
```

```python
usa1965.income.plot(kind = 'hist', histtype = 'step', bins = 30)
plt.axvline(usa1965.income.mean())
plt.axvline(usa1965.income.median(), linestyle = '--')
plt.axvline(usa1965.income.quantile(0.25), linestyle = ':')
plt.axvline(usa1965.income.quantile(0.75), linestyle = ':')
```

```python
usa1965.income.plot(kind = 'hist', histtype = 'step', bins = 30, density = True)
usa1965.income.plot.density()
```

```python
ax = china1965.plot.scatter('income', 'log10_income')
china1965.income.plot.line(x = 'income', y = 'log10_income', ax = ax)
```
