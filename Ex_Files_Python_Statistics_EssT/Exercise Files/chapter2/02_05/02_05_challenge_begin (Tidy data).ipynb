```python
import numpy as np
import pandas as pd

import matplotlib
import matplotlib.pyplot as pp

%matplotlib inline
```

```python
tb = pd.read_csv('tb.csv')
tb.head()
```

```python
df = pd.melt(tb, id_vars = ['country', 'year'], var_name = 'age', value_name = 'cases')
df.head()
```

```python
df['sex'], df['age_range'] = df['age'].str[0:1], df['age'].str[1:]
df.head()
```

```python
df.info()
```

```python
df = df.dropna()
df.head()
```

```python
df.drop(['age'], axis = 1)
```

```python
df.sort_values(['country', 'year', 'age'])
```

```python
final = df[['country', 'year', 'age_range', 'sex', 'cases']]
final.head()
```

```python
final.to_csv('final_judith.csv', index = False)
```
