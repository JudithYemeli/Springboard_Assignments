 ## Role of the cursor object
 acts like a pointer to a specific row in a query result set
 How Cursors Work:
Declare: You declare a cursor by specifying a SQL query that retrieves the desired result set.
Open: The cursor is then opened, which executes the query and makes the result set available.
Fetch: You can then fetch rows one at a time (or in bulk, depending on the cursor type) and process each row as needed.
Close: Once done, you close the cursor to release the resources.
Deallocate: Finally, you deallocate the cursor to remove it from memory.

```python
q5 = """WITH CastCounts AS (
    SELECT 
        cast_id,
        cast_name,
        COUNT(*) AS appearances
    FROM movie_cast
    GROUP BY cast_id, cast_name
    HAVING COUNT(*) > 3
),
CastAvgScores AS (
    SELECT 
        cc.cast_id, 
        cc.cast_name, 
        AVG(m.score) AS avg_score
    FROM CastCounts cc
    INNER JOIN movie_cast mc ON cc.cast_id = mc.cast_id
    INNER JOIN moviex m ON mc.movie_id = m.id
    WHERE m.score < 25
    GROUP BY cc.cast_id, cc.cast_name
)
SELECT 
    cast_id, 
    cast_name, 
    avg_score
FROM CastAvgScores
ORDER BY avg_score DESC, cast_name ASC;"""
```

### db partitionng 

####why? 
Improve speed/performance because only a subset of the data is query
Preserve data integrity (sensitive columns can be isolated)

#### what?
vertical partitioning: splitting the columns
horizontal partitioning: splitting the rows

#### How (vertical partitioning)?
1. create the original table to be partitioned (a primary key should be identified)
2. Determine the split of columns
3. create the 2 partitioned tables. The columns names need to be the same as the original
    (primary key should also be identified and the same as in the original)
4. Modify the structure of the EmployeeDetails table to add a fore ign key constraint
ALTER TABLE EmployeeDetails:

This statement specifies that you are modifying the structure of the EmployeeDetails table.
ADD CONSTRAINT FK_Employee_EmployeeID:

This part of the statement is adding a new constraint to the EmployeeDetails table. The constraint is named FK_Employee_EmployeeID. Naming constraints helps identify them easily for future reference, such as during maintenance or troubleshooting.
FOREIGN KEY (EmployeeID):

This specifies that the EmployeeID column in the EmployeeDetails table will be a foreign key. A foreign key is a column (or set of columns) in one table that uniquely identifies a row of another table.
REFERENCES EmployeeBasicInfo(EmployeeID):

This indicates that the EmployeeID column in the EmployeeDetails table is a foreign key that references the EmployeeID column in the EmployeeBasicInfo table. This establishes a relationship between the two tables.

q7 - Creating views. Create a view (virtual table) called good_collaboration that lists pairs of actors who have had a good collaboration as defined here. Each row in the view describes one pair of actors who appeared in at least 2 movies together AND the average score of these movies is >= 40.
The view should have the format:
     good_collaboration(
           cast_member_id1,
           cast_member_id2,
           movie_count,
              average_movie_score)

```python
q7 = """create view good_collaboration as
        select 
                c1.cast_id as cast_member_id1, 
                c2.cast_id as cast_member_id2, 
                count(*) as movie_count
                avg(score) as average_movie_score
        from 
                movie_cast as c1
        cross join movie_cast as c2
        inner join movies on movie_id = id
        where cast_member_id1 < cast_member_id2
        group by cast_member_id1, cast_member_id2
        having movie_count >= 2 and average_movie_score >=40
                """
```
